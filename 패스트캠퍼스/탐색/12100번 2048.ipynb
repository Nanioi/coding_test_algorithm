```python
from copy import deepcopy

N = int(input())
Board = [list(map(int,input().split())) for i in range(N)]

# 90도  돌리기
def rotate90(B,N):
    NB = deepcopy(B)
    for i in range(N):
        for j in range(N):
            NB[j][N-i-1]=B[i][j]
    return NB 

# 왼쪽으로 슬라이드 시 생기는 결과값
def convert(lst,N):
    new_list = [i for i in lst if i] #양수만 복사
    for i in range(1,len(new_list)):
        if new_list[i-1] == new_list[i]:
            new_list[i-1]*=2
            new_list[i]=0
    new_list = [i for i in new_list if i]
    return new_list + [0] *(N-len(new_list)) #나머지 0으로 채우고 리턴

def dfs(N,B,count):
    ret = max([max(i) for i in B])
    if count==0:
        return ret
    for _ in range(4):
        X = [convert(i,N) for i in B]
        if X != B:
            ret = max(ret,dfs(N,X,count-1))
        B = rotate90(B,N)
    return ret



print(dfs(N,Board,5))
```

```python
from copy import deepcopy

n = int(input())
M = [list(map(int,input().split())) for _ in range(n)]

def rotate90(m):
    board = deepcopy(m)
    for i in range(n):
        for j in range(n):
            board[j][n-i-1] = m[i][j]
    return board

def convert(lst):
    new_list = [i for i in lst if i] # 양수만 복사
    for i in range(1,len(new_list)):
        if new_list[i-1] == new_list[i]:
            new_list[i-1]*=2
            new_list[i]=0
    new_list = [i for i in new_list if i]
    return new_list + [0]*(n-len(new_list))
            
def dfs(n,m,cnt):
    ret = max([max(i) for i in m])
    if cnt==0 : 
        return ret
    for _ in range(4):
        x = [convert(i) for i in m] ## 왼쪽 슬라이드 시 결과
        if x != m:
            ret = max(ret, dfs(n,x,cnt-1))
        m = rotate90(m) # 90도 돌리기
    return ret

print(dfs(n,M,5))
```
