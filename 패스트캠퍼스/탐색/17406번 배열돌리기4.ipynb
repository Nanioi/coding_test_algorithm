```python
#bitmask

from copy import deepcopy

N,M,K = map(int,input().split())
A=[list(map(int,input().split())) for _ in range(N)]
Q=[tuple(map(int,input().split()))for _ in range(K)]

dx,dy = [1,0,-1,0],[0,-1,0,1]

def value(arr):
    return min([sum(i) for i in arr])

def convert(arr,qry):
    (r,c,s)=qry
    r,c = r-1,c-1
    new_arr = deepcopy(arr)

    for i in range(1,s+1):
        rr,cc = r-i,c+i
        for w in range(4):
            for _ in range(i*2):
                rrr,ccc = rr+dx[w] , cc+dy[w]
                new_arr[rrr][ccc]=arr[rr][cc]
                rr,cc = rrr,ccc
    return new_arr

def dfs(arr,qry,st):
    if st == (1<<K)-1: return value(arr) # 쿼리 처리 다한 경우 
    ret = 50000
    for i in range(K):
        if st&1<<i==0:ret=min(ret,dfs(convert(arr,qry[i]),qry,st|1<<i)) # 처리 안한 경우
    return ret

print(dfs(A,Q,0))
```
