```python
def solution(distance, rocks, n):
    answer = 0
    sorted_rocks = sorted(rocks)
    sorted_rocks.append(distance)
    left = 0
    right = distance
    while (left <= right):
        mid = int((left + right) / 2)
        removeRockCnt = 0
        prev = 0
        for i in range(len(sorted_rocks)):
            if (sorted_rocks[i] - prev < mid):
                removeRockCnt += 1
            else:
                prev = sorted_rocks[i]
        if removeRockCnt > n:
            right = mid - 1
        else:
            left = mid + 1
            answer = mid
    return answer
```

```python
def solution(distance, rocks, n):
    rocks.sort()
    low, high = 0, distance
    while high > low:
        dist = (low + high + 1) // 2
        prev = hits = 0
        for i in range(len(rocks)):
            if rocks[i] - prev < dist: hits += 1
            else: prev = rocks[i]
        if hits > n: high = dist - 1
        else: low = dist
    return high
```
